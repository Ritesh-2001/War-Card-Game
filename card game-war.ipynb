```python
import random
```

```python
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':11, 'Queen':12, 'King':13, 'Ace':14}
```

```python
class Card:
    def __init__(self,suit,rank):
        self.suit=suit
        self.rank=rank
        self.value=values[self.rank]
        
    def __str__(self):                         #eg-print('<two_hearts>')
        return self.rank+' of '+self.suit      #output-two of hearts
```

```python
class Deck:
    def __init__(self):
        # Note this only happens once upon creation of a new Deck
        self.all_cards = [] #A list
        for suit in suits:       #creating a new deck of cards
            for rank in ranks:
                created_card=Card(suit,rank)   #creating a card object
                # This assumes the Card class has already been defined!
                self.all_cards.append(created_card)
                
    def shuffle(self):
        random.shuffle(self.all_cards)      #shuffles list self.all_cards
       
    def deal_one(self):
        # Note we remove one card from the list of all_cards
        return self.all_cards.pop()   
```

```python
class Player:
    
    def __init__(self,name):
        self.name = name
        # A new player has no cards
        self.all_cards = [] 
        
    def remove_one(self):
        # Note we remove one card from the list of all_cards
        # We state 0 to remove from the "top" of the deck
        # We'll imagine index -1 as the bottom of the deck
        return self.all_cards.pop(0) #remove first card in deck
    
    def add_cards(self,new_cards):
        if type(new_cards) == type([]):
            self.all_cards.extend(new_cards)   #extend when we are adding multiple cards,i.e new_cards is a  
        else:
            self.all_cards.append(new_cards)   #append when we are adding single card
    
    
    def __str__(self):
        return f'Player {self.name} has {len(self.all_cards)} cards.'
```

```python
player_one = Player("One")
player_two = Player("Two")
```
